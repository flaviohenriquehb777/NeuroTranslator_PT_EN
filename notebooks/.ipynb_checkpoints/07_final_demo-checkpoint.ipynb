## 1. Sistema de DemonstraÃ§Ã£o Interativa

````python
class NeuroTranslatorDemo:
    """
    Sistema de demonstraÃ§Ã£o do NeuroTranslator
    """
    
    def __init__(self):
        self.version = "1.0.0"
        self.is_running = False
        self.demo_data = self._load_demo_data()
        self.metrics = {
            'translations': 0,
            'audio_processed': 0,
            'video_frames': 0,
            'start_time': None,
            'latencies': [],
            'languages_detected': {'pt': 0, 'en': 0}
        }
        
        print(f"ğŸ§  NeuroTranslator v{self.version} inicializado!")
        print(f"ğŸ“Š {len(self.demo_data)} frases de demonstraÃ§Ã£o carregadas")
    
    def _load_demo_data(self):
        """
        Carrega dados de demonstraÃ§Ã£o
        """
        return [
            # ConversaÃ§Ã£o bÃ¡sica
            {
                'category': 'ConversaÃ§Ã£o BÃ¡sica',
                'pt': 'OlÃ¡, como vocÃª estÃ¡ hoje?',
                'en': 'Hello, how are you today?',
                'context': 'Cumprimento casual'
            },
            {
                'category': 'ConversaÃ§Ã£o BÃ¡sica',
                'pt': 'Muito bem, obrigado! E vocÃª?',
                'en': 'Very well, thank you! And you?',
                'context': 'Resposta educada'
            },
            {
                'category': 'ConversaÃ§Ã£o BÃ¡sica',
                'pt': 'Prazer em conhecÃª-lo!',
                'en': 'Nice to meet you!',
                'context': 'ApresentaÃ§Ã£o'
            },
            
            # ReuniÃµes de trabalho
            {
                'category': 'ReuniÃµes',
                'pt': 'Vamos comeÃ§ar a reuniÃ£o de hoje.',
                'en': "Let's start today's meeting.",
                'context': 'InÃ­cio de reuniÃ£o'
            },
            {
                'category': 'ReuniÃµes',
                'pt': 'Qual Ã© o status do projeto atual?',
                'en': 'What is the status of the current project?',
                'context': 'Pergunta sobre projeto'
            },
            {
                'category': 'ReuniÃµes',
                'pt': 'Precisamos revisar os prazos estabelecidos.',
                'en': 'We need to review the established deadlines.',
                'context': 'DiscussÃ£o de cronograma'
            },
            
            # SituaÃ§Ãµes tÃ©cnicas
            {
                'category': 'TÃ©cnico',
                'pt': 'O sistema estÃ¡ funcionando corretamente?',
                'en': 'Is the system working correctly?',
                'context': 'VerificaÃ§Ã£o tÃ©cnica'
            },
            {
                'category': 'TÃ©cnico',
                'pt': 'Vamos implementar uma nova funcionalidade.',
                'en': 'We are going to implement a new feature.',
                'context': 'Desenvolvimento'
            },
            {
                'category': 'TÃ©cnico',
                'pt': 'Os testes foram executados com sucesso.',
                'en': 'The tests were executed successfully.',
                'context': 'Resultado de testes'
            },
            
            # SituaÃ§Ãµes cotidianas
            {
                'category': 'Cotidiano',
                'pt': 'Que horas sÃ£o agora?',
                'en': 'What time is it now?',
                'context': 'Pergunta sobre horÃ¡rio'
            },
            {
                'category': 'Cotidiano',
                'pt': 'Onde fica a estaÃ§Ã£o de trem mais prÃ³xima?',
                'en': 'Where is the nearest train station?',
                'context': 'Pedindo direÃ§Ãµes'
            },
            {
                'category': 'Cotidiano',
                'pt': 'Obrigado pela sua ajuda!',
                'en': 'Thank you for your help!',
                'context': 'Agradecimento'
            },
            
            # ExpressÃµes complexas
            {
                'category': 'Complexo',
                'pt': 'Estou muito animado com as possibilidades que este projeto oferece.',
                'en': 'I am very excited about the possibilities this project offers.',
                'context': 'ExpressÃ£o de entusiasmo'
            },
            {
                'category': 'Complexo',
                'pt': 'Precisamos analisar cuidadosamente todos os aspectos antes de tomar uma decisÃ£o.',
                'en': 'We need to carefully analyze all aspects before making a decision.',
                'context': 'AnÃ¡lise detalhada'
            },
            {
                'category': 'Complexo',
                'pt': 'A implementaÃ§Ã£o desta soluÃ§Ã£o pode revolucionar nossa abordagem atual.',
                'en': 'The implementation of this solution could revolutionize our current approach.',
                'context': 'Impacto transformador'
            }
        ]
    
    def start_demo(self, duration_minutes=5, interactive=True):
        """
        Inicia demonstraÃ§Ã£o do sistema
        """
        print(f"\nğŸ¬ Iniciando demonstraÃ§Ã£o por {duration_minutes} minutos...")
        print("ğŸ¯ Modo:", "Interativo" if interactive else "AutomÃ¡tico")
        print("="*50)
        
        self.is_running = True
        self.metrics['start_time'] = datetime.now()
        
        if interactive:
            self._interactive_demo(duration_minutes)
        else:
            self._automatic_demo(duration_minutes)
        
        self._show_final_report()
    
    def _interactive_demo(self, duration_minutes):
        """
        DemonstraÃ§Ã£o interativa
        """
        print("ğŸ® DemonstraÃ§Ã£o Interativa Iniciada!")
        print("ğŸ’¡ Dica: O sistema processarÃ¡ frases automaticamente")
        print("â±ï¸ Pressione Ctrl+C para parar a qualquer momento\n")
        
        end_time = datetime.now() + timedelta(minutes=duration_minutes)
        phrase_index = 0
        
        try:
            while datetime.now() < end_time and self.is_running:
                # Selecionar frase
                phrase = self.demo_data[phrase_index % len(self.demo_data)]
                phrase_index += 1
                
                # Decidir direÃ§Ã£o da traduÃ§Ã£o
                pt_to_en = np.random.choice([True, False])
                
                if pt_to_en:
                    self._process_translation(
                        phrase['pt'], phrase['en'], 'pt', 'en', phrase
                    )
                else:
                    self._process_translation(
                        phrase['en'], phrase['pt'], 'en', 'pt', phrase
                    )
                
                # Intervalo entre traduÃ§Ãµes
                time.sleep(np.random.uniform(3, 8))
                
        except KeyboardInterrupt:
            print("\nâ¹ï¸ DemonstraÃ§Ã£o interrompida pelo usuÃ¡rio")
        
        self.is_running = False
    
    def _automatic_demo(self, duration_minutes):
        """
        DemonstraÃ§Ã£o automÃ¡tica
        """
        print("ğŸ¤– DemonstraÃ§Ã£o AutomÃ¡tica Iniciada!")
        print("ğŸ“Š Processando todas as categorias de frases\n")
        
        # Agrupar por categoria
        categories = {}
        for phrase in self.demo_data:
            cat = phrase['category']
            if cat not in categories:
                categories[cat] = []
            categories[cat].append(phrase)
        
        # Processar cada categoria
        for category, phrases in categories.items():
            print(f"\nğŸ“‚ Categoria: {category}")
            print("-" * 40)
            
            for phrase in phrases:
                # Traduzir em ambas as direÃ§Ãµes
                self._process_translation(
                    phrase['pt'], phrase['en'], 'pt', 'en', phrase
                )
                time.sleep(1)
                
                self._process_translation(
                    phrase['en'], phrase['pt'], 'en', 'pt', phrase
                )
                time.sleep(1)
            
            time.sleep(2)  # Pausa entre categorias
        
        self.is_running = False
    
    def _process_translation(self, source_text, target_text, src_lang, tgt_lang, phrase_data):
        """
        Processa uma traduÃ§Ã£o completa
        """
        start_time = time.time()
        timestamp = datetime.now().strftime("%H:%M:%S")
        
        print(f"\nğŸ”„ [{timestamp}] Processando traduÃ§Ã£o...")
        print(f"ğŸ“ Contexto: {phrase_data['context']}")
        print(f"ğŸŒ DireÃ§Ã£o: {src_lang.upper()} â†’ {tgt_lang.upper()}")
        
        # Simular pipeline completo
        pipeline_steps = [
            ("ğŸ¤ Captura de Ã¡udio", np.random.uniform(50, 150)),
            ("ğŸ—£ï¸ Speech-to-text", np.random.uniform(100, 300)),
            ("ğŸŒ DetecÃ§Ã£o de idioma", np.random.uniform(10, 50)),
            ("ğŸ”„ TraduÃ§Ã£o neural", np.random.uniform(200, 500)),
            ("ğŸ”Š Text-to-speech", np.random.uniform(300, 800)),
            ("ğŸ“º ExibiÃ§Ã£o de legendas", np.random.uniform(20, 100)),
            ("ğŸ“¹ Processamento de vÃ­deo", np.random.uniform(30, 100))
        ]
        
        total_latency = 0
        
        for step_name, step_time in pipeline_steps:
            time.sleep(step_time / 1000)  # Simular processamento
            total_latency += step_time
            print(f"   {step_name}: {step_time:.0f}ms")
        
        # Mostrar textos
        print(f"\nğŸ“ Original ({src_lang}): '{source_text}'")
        print(f"ğŸ“ Traduzido ({tgt_lang}): '{target_text}'")
        
        # Atualizar mÃ©tricas
        self.metrics['translations'] += 1
        self.metrics['audio_processed'] += 1
        self.metrics['video_frames'] += np.random.randint(30, 60)
        self.metrics['latencies'].append(total_latency)
        self.metrics['languages_detected'][src_lang] += 1
        
        # Calcular qualidade simulada
        quality_score = np.random.uniform(0.85, 0.98)
        confidence = np.random.uniform(0.90, 0.99)
        
        print(f"\nğŸ“Š MÃ©tricas:")
        print(f"   âš¡ LatÃªncia total: {total_latency:.0f}ms")
        print(f"   ğŸ¯ Qualidade: {quality_score:.2f}")
        print(f"   ğŸ”’ ConfianÃ§a: {confidence:.2f}")
        print(f"   ğŸ¯ TraduÃ§Ã£o #{self.metrics['translations']}")
    
    def _show_final_report(self):
        """
        Mostra relatÃ³rio final da demonstraÃ§Ã£o
        """
        if not self.metrics['start_time']:
            return
        
        end_time = datetime.now()
        duration = end_time - self.metrics['start_time']
        
        print("\n" + "="*60)
        print("ğŸ“Š RELATÃ“RIO FINAL DA DEMONSTRAÃ‡ÃƒO")
        print("="*60)
        
        print(f"\nâ±ï¸ TEMPO DE EXECUÃ‡ÃƒO:")
        print(f"   InÃ­cio: {self.metrics['start_time'].strftime('%H:%M:%S')}")
        print(f"   Fim: {end_time.strftime('%H:%M:%S')}")
        print(f"   DuraÃ§Ã£o: {duration}")
        
        print(f"\nğŸ“ˆ ESTATÃSTICAS:")
        print(f"   ğŸ”„ TraduÃ§Ãµes realizadas: {self.metrics['translations']}")
        print(f"   ğŸ¤ Ãudios processados: {self.metrics['audio_processed']}")
        print(f"   ğŸ“¹ Frames de vÃ­deo: {self.metrics['video_frames']}")
        
        if self.metrics['latencies']:
            avg_latency = np.mean(self.metrics['latencies'])
            min_latency = np.min(self.metrics['latencies'])
            max_latency = np.max(self.metrics['latencies'])
            
            print(f"\nâš¡ PERFORMANCE:")
            print(f"   LatÃªncia mÃ©dia: {avg_latency:.0f}ms")
            print(f"   LatÃªncia mÃ­nima: {min_latency:.0f}ms")
            print(f"   LatÃªncia mÃ¡xima: {max_latency:.0f}ms")
        
        if duration.total_seconds() > 0:
            rate = self.metrics['translations'] / duration.total_seconds() * 60
            print(f"   ğŸš€ Taxa: {rate:.1f} traduÃ§Ãµes/minuto")
        
        print(f"\nğŸŒ IDIOMAS PROCESSADOS:")
        for lang, count in self.metrics['languages_detected'].items():
            lang_name = "PortuguÃªs" if lang == 'pt' else "InglÃªs"
            print(f"   {lang_name}: {count} frases")
        
        print("\n" + "="*60)
        print("âœ… DemonstraÃ§Ã£o concluÃ­da com sucesso!")
        print("ğŸ‰ NeuroTranslator funcionando perfeitamente!")
    
    def plot_performance_metrics(self):
        """
        Plota mÃ©tricas de performance
        """
        if not self.metrics['latencies']:
            print("âš ï¸ Nenhuma mÃ©trica disponÃ­vel para plotar")
            return
        
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle('ğŸ“Š MÃ©tricas de Performance - NeuroTranslator', fontsize=16)
        
        # LatÃªncias ao longo do tempo
        axes[0, 0].plot(self.metrics['latencies'], 'b-o', markersize=4)
        axes[0, 0].set_title('âš¡ LatÃªncia por TraduÃ§Ã£o')
        axes[0, 0].set_xlabel('TraduÃ§Ã£o #')
        axes[0, 0].set_ylabel('LatÃªncia (ms)')
        axes[0, 0].grid(True, alpha=0.3)
        
        # Histograma de latÃªncias
        axes[0, 1].hist(self.metrics['latencies'], bins=15, alpha=0.7, color='green')
        axes[0, 1].set_title('ğŸ“Š DistribuiÃ§Ã£o de LatÃªncias')
        axes[0, 1].set_xlabel('LatÃªncia (ms)')
        axes[0, 1].set_ylabel('FrequÃªncia')
        axes[0, 1].grid(True, alpha=0.3)
        
        # Idiomas processados
        languages = list(self.metrics['languages_detected'].keys())
        counts = list(self.metrics['languages_detected'].values())
        lang_names = ['PortuguÃªs' if l == 'pt' else 'InglÃªs' for l in languages]
        
        axes[1, 0].pie(counts, labels=lang_names, autopct='%1.1f%%', startangle=90)
        axes[1, 0].set_title('ğŸŒ DistribuiÃ§Ã£o de Idiomas')
        
        # MÃ©tricas gerais
        metrics_names = ['TraduÃ§Ãµes', 'Ãudios', 'Frames']
        metrics_values = [
            self.metrics['translations'],
            self.metrics['audio_processed'],
            self.metrics['video_frames']
        ]
        
        bars = axes[1, 1].bar(metrics_names, metrics_values, color=['blue', 'orange', 'red'])
        axes[1, 1].set_title('ğŸ“ˆ MÃ©tricas Gerais')
        axes[1, 1].set_ylabel('Quantidade')
        
        # Adicionar valores nas barras
        for bar, value in zip(bars, metrics_values):
            height = bar.get_height()
            axes[1, 1].text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                           f'{value}', ha='center', va='bottom')
        
        plt.tight_layout()
        plt.show()

print("ğŸ¬ Sistema de demonstraÃ§Ã£o implementado!")
````
